```python
# Импортируем библиотеки
import pandas as pd

# Загружаем набор данных
ECommerce_Shipping_Data = pd.read_csv('/content/drive/MyDrive/ECommerce_Shipping_Data.csv')
ECommerce_Shipping_Data.head()
```

```python
# Сохраняем в новый фрейм данных столбцы предикторов
X=ECommerce_Shipping_Data.drop(['Reached_on_Time_Y_N'], axis=1)
# Сохраняем в переменную y отклик
y=ECommerce_Shipping_Data.Reached_on_Time_Y_N
```

```python
# Преобразуем категориальные переменные в dummy
X=pd.get_dummies(X, drop_first=True)
```

```python
from sklearn.linear_model import LogisticRegression
lorg = LogisticRegression()
lorg.fit(X, y)
```

```python
lorg.coef_
```

```python
lorg.intercept_
```

```python
lorg.score(X, y)
```

```python
import numpy as np
np.array(lorg.coef_).flatten()
```

```python
klogr = pd.DataFrame()
klogr['name']=X.columns
klogr['coef']=np.array(lorg.coef_).flatten()
klogr
```

```python
lorg.predict(X)
```
